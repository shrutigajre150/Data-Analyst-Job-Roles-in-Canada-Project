# Data Analyst Job Roles in Canada

## About Dataset
- Welcome to a comprehensive dataset of Data Analyst job roles across Canada!
- This dataset provides a unique glimpse into the job market, capturing essential details like salary ranges, required skills, programming languages, job titles, employers, and much more.



### Featured Columns in the Clean Dataset:
- Job Title: A generalized job title that encapsulates the role.
- Job Info: The exact job title as listed on the job sites.
- Position: The specific role or category the job falls under.
- Employer: The name of the hiring company.
- City: The job's location.
- Province: The abbreviated province name corresponding to the city.
- Skill: The programming languages and tools required for the job.
- Seniority: The job's seniority level (Senior, Mid, Junior, any).
- Work Type: Specifies if the job is Remote, In-person, or Hybrid.
- Industry Type: The industry to which the employer belongs.
- Min Salary: The lowest salary listed (as a float).
- Max Salary: The highest salary listed (as a float).
- Average Salary: The mean salary (as a float).

#### 3 What is the salary range (Min_Salary, Max_Salary) for each job title?

```python
salary_range = df.groupby('Job Title')["Min_Salary", "Max_Salary"].agg(['min', 'max'])
```

#### 4 Which skills are associated with higher average salaries?

```python
avg_salary_by_skill = df.groupby('Skill')['Avg_Salary'].mean()
sorted_avg_salary_by_skill = avg_salary_by_skill.sort_values(ascending=False)
```

```python
avg_salary_by_skill
```

```python
sorted_avg_salary_by_skill
```

#### 5. Which job titles show the highest growth potential in terms of salary and job availability?

```python
df['Salary_Growth'] = df['Max_Salary'] - df['Min_Salary']

job_analysis = df.groupby('Job Title').agg(
    Avg_Salary=('Avg_Salary', 'mean'),
    Salary_Growth=('Salary_Growth', 'mean'),
    Job_Availability=('Job Title', 'size')
)
job_analysis_sorted = job_analysis.sort_values(by=['Salary_Growth', 'Job_Availability'], ascending=False)
job_analysis_sorted 
```

```python
top_5_skills = df.groupby('Skill')['Min_Salary'].mean().nlargest(5)
top_5_skills 
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_skills.values, y=top_5_skills.index, palette="viridis")
plt.xlabel("Average Minimum Salary")
plt.ylabel("Skill")
plt.title("Top 5 Skills by Average Minimum Salary")
plt.show()
```

```python
top_5_skills_Max = df.groupby('Skill')['Max_Salary'].mean().nlargest(5)
top_5_skills_Max  
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_skills_Max.values, y=top_5_skills_Max.index, palette="viridis")
plt.xlabel("Average Minimum Salary")
plt.ylabel("Skill")
plt.title("Top 5 Skills by Average Minimum Salary")
plt.show()
```

```python
# Distribution of Average Salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Avg_Salary'], bins=20, kde=True, color="steelblue")
plt.xlabel("Average Salary")
plt.title("Distribution of Average Salaries")
plt.show()
```

```python
# Distribution of Min Salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Min_Salary'], bins=20, kde=True, color="yellowgreen")
plt.xlabel("Min Salary")
plt.title("Distribution of Min Salaries")
plt.show()
```

```python
# Distribution of Max Salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Max_Salary'], bins=20, kde=True, color="green")
plt.xlabel("Max Salary")
plt.title("Distribution of Max Salaries")
plt.show()
```

```python
# Min vs. Max Salary Comparison by Industry Type

industry_salary = df.groupby('Industry Type')[['Min_Salary', 'Max_Salary']].mean().nlargest(10, 'Max_Salary')
industry_salary.plot(kind='bar', figsize=(12, 8), color=['skyblue', 'Thistle'])
plt.ylabel("Salary")
plt.xlabel("Industry Type")
plt.title("Min vs Max Salary by Industry Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Positions by Average Salary
top_positions = df.groupby('Position')['Avg_Salary'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_positions.values, y=top_positions.index, palette="Paired")
plt.xlabel("Average Salary")
plt.title("Top 10 Positions by Average Salary")
plt.show()
```
